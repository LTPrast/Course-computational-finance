```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde
import seaborn as sns
sns.set_theme()
import os

SAVE_DIR = "data"

if not os.path.exists(SAVE_DIR):
    os.mkdir(SAVE_DIR)
    
Φ = norm.cdf

def fst(t):
    return t[0]

def snd(t):
    return t[1]

def lst(x):
    return x[len(x) - 1]

def sample_mean(xs):
    return np.mean(xs)

def sample_variance(xs):
    N = len(xs)
    μ = sample_mean(xs)
    σsq = np.sum((xs-μ)**2)/(N-1)
    return σsq

def sample_mean_variance(xs):
    N = len(xs)
    μ = sample_mean(xs)
    σsq = np.sum((xs-μ)**2)/(N-1)
    return μ, σsq

def discrete_derivative_kernel(xs):
    """
    Calculates the series y[i] = x[i+1] - x[i]
    """
    N = len(xs)
    dxs = np.zeros(N-1)
    for i in range(N-1):
        dxs[i] = xs[i+1] - xs[i]
    return dxs
```

```python
def save_path(filename):
    if SAVE_DIR[-1] == '/':
        return f"{SAVE_DIR}{filename}"
    else:
        return f"{SAVE_DIR}/{filename}"
```

```python
# def call_option_payoff(asset_price, strike_price):
#     """
#     Payoff (S - K)^+
#     """
#     return max(asset_price - strike_price, 0.0)

def d1(t, T, St, K, r, σ):
    # checked
    τ = T-t
    numerator = np.log(St/K) + (r + 0.5 * np.power(σ, 2)) * τ
    denominator = σ*np.sqrt(τ)
    return numerator/denominator

def d2(t, T, St, K, r, σ):
    # checked
    τ = T-t
    return d1(t, T, St, K, r, σ) - σ*np.sqrt(τ)

def Δ(t, T, St, K, r, σ):
    """
    Analytical form of Δ(t) as given by the Black-Scholes model
    """
    return Φ(d1(t, T, St, K, r, σ))

def black_scholes_GBM_valuation(t, T, St, K, r, σ):
    """
    V(t) as given by the Black-Scholes formula for asset price exhibiting GBM
    """
    # checked
    τ = T - t
    rτ = r*τ
    d1_value = d1(t, T, St, r, K, σ)
    d2_value = d2(t, T, St, r, K, σ)
    return St*Φ(d1_value) - np.exp(-rτ)*K*Φ(d2_value)

def initial_bond_value(T, S0, r, K, σ):
    Δ0 = Δ(0, T, S0, K, r, σ)
    initial_equity = S0 * Δ0
    option_price = black_scholes_GBM_valuation(0, T, S0, K, r, σ)
    return option_price - initial_equity

def euler_GBM(adjustment_frequency, initial_stock_price, strike_price, maturity_time, rf_interest_rate, σ, n_sim, σ_Δ = None):
    if not σ_Δ:
        σ_Δ = σ
    T = maturity_time
    M = adjustment_frequency
    S0 = initial_stock_price
    K = strike_price
    r = rf_interest_rate
    
    # allocate mem for asset prices
    S_timeseries = np.zeros(shape=(n_sim, np.uint64(M+1)))  # we need this type-conversion here
    # set initial asset prices to S0
    S_timeseries[:,0] = S0
    
    # allocate mem for Δs
    Δ_values = np.zeros(shape=S_timeseries.shape)
    # set initial value of Δ
    Δ_values[:,0] = Δ(0, T, S0, K, r, σ_Δ)
    
    # allocate mem for bond values
    bond_values = np.zeros(shape=S_timeseries.shape)
    # set initial value of bond values
    bond_values[:,0] = initial_bond_value(T, S0, r, K, σ)
    
    # allocate mem for option values
    option_values = np.zeros(shape=S_timeseries.shape)
    # set initial option value
    initial_option_value = black_scholes_GBM_valuation(0,T, S0, K, r, σ)
    option_values[:,0] = initial_option_value
    dt = T/M
        
    # run multiple simulations
    for sim in range(n_sim):
        # simulate the asset price
        # calculate the Δ values
        for m in range(M):
            t = (m+1)*dt
            zm = np.random.normal(loc=0.0, scale=1.0)
            St = S_timeseries[sim][m] * (1.0 + r*dt  + σ*np.sqrt(dt)*zm)
            S_timeseries[sim][m+1] = St
            Δ_values[sim][m+1] = Δ(t, T, St, K, r, σ_Δ)
            option_values[sim][m+1] = black_scholes_GBM_valuation(t, T, St, K, r, σ)
        dΔ_values = discrete_derivative_kernel(Δ_values[sim])
        # perform the hedging for this timeseries
        for i in range(len(dΔ_values)):
            # the previous timestep earns continuously compounded interest
            bond_values[sim][i+1] = bond_values[sim][i] * np.exp(r*dt)
            # part of bond used to hedge
            bond_values[sim][i+1] -= dΔ_values[i] * S_timeseries[sim][i+1]
    
    return S_timeseries, Δ_values, bond_values, option_values

def portfolio_value(asset_qty, asset_price, bond_value, option_value):
    equity = asset_price * asset_qty
    return equity + bond_value - option_value
```

```python
adjustment_frequencies = np.array([365] + list(reversed(range(150, 301, 50))) + [52])
assert len(adjustment_frequencies) == 6
T = 1.0
σ = 0.2
S0 = 100.0
K = 99.0
n_sim = 10
r = 0.06
data = euler_GBM(lst(adjustment_frequencies), S0, K, T, r, σ, n_sim)
S_data, Δ_data, B_data, C_data = data
equity_data = Δ_data * S_data
Π_data = S_data * Δ_data + B_data - C_data
final_portfolio_values = np.array(list(map(lst, Π_data)))
```

```python
def sim_iter_adjustment_frequencies(Ms, T, S0, K, r, σ, n_sim, σ_Δ=None, v=False):
    final_portfolio_data = []
    for M in Ms:
        if v:
            print(f"Simulating {n_sim} iterations for M = {M}")
        S_data, Δ_data, B_data, C_data = euler_GBM(M, S0, K, T, r, σ, n_sim, σ_Δ = σ_Δ)
        # def euler_GBM(adjustment_frequency, initial_stock_price, strike_price, maturity_time, rf_interest_rate, σ, n_sim, σ_Δ = None):
        Π_data = portfolio_value(Δ_data, S_data, B_data, C_data)
        final_portfolio_values = np.array(list(map(lst, Π_data)))
        final_portfolio_data.append(final_portfolio_values)
    
    return final_portfolio_data
```

```python
def make_histplots(M_values, final_portfolio_data):
    fig, axes = plt.subplots(2, 3, figsize=(16, 10), sharey=False, sharex=True)
    fig.suptitle("Final Portfolio Values")
    for i,axrow in enumerate(axes):
        row = i*snd(M_values.shape)
        for j,ax in enumerate(axrow):
            ax.set_title(f"M = {M_values[i][j]}")
            ax.set_ylabel("$\\Pi(t)$")
            sns.histplot(ax=ax,data=final_portfolio_data[row+j],bins=50, kde=True)
    return fig, axes
```

```python
group_plot_data = []
adjustment_frequencies = np.array([365] + list(reversed(range(150, 301, 50))) + [52])
M_values = adjustment_frequencies.reshape((2,3))
assert len(adjustment_frequencies) == 6
T = 1.0
σ = 0.2
S0 = 100.0
K = 99.0
n_sim = 1000
r = 0.06

npz_filenames = {M: f"matched_volatility_{M}.npz" for M in adjustment_frequencies}
npz_savepaths = {M: f"{SAVE_DIR}/{npz_filenames[M]}" for M in npz_filenames.keys()}

final_portfolio_data = []
if not os.path.exists(npz_savepaths[lst(adjustment_frequencies)]):
    final_portfolio_data = sim_iter_adjustment_frequencies(adjustment_frequencies, T, S0, K, r, σ, n_sim, v=True)
    for M, Π_data in zip(adjustment_frequencies, final_portfolio_data):
        save_path = npz_savepaths[M]
        np.savez(save_path, Π_data)
else:
    for M in adjustment_frequencies:
        with np.load(npz_savepaths[M]) as f:
            final_portfolio_data.append(f['arr_0'])
            
make_histplots(M_values, final_portfolio_data)
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 15 16:41:43 2022

@author: LPras
"""
import numpy as np
import matplotlib.pyplot as plt
from utils import black_scholes, FTCS, CN
import seaborn as sns
import pandas as pd


def PDE(S_0, K, r, sigma, T, t_steps, x_steps,dt,scheme):
 
    x_max = np.log(S_0) + 20*sigma*np.sqrt(T)
    x_min = np.log(S_0) - 20*sigma*np.sqrt(T)
    dx = (x_max - x_min)/x_steps
    
    #everly spaced logarithacly
    X = np.linspace(x_min, x_max, x_steps+1)
    #then add the the exponential
    S = np.exp(X)
    Smax = S[-1]
    #print(S)
    

    #initialize grid and (at Xmin/Smin -> V=0)
    V = np.zeros((x_steps+1,t_steps+1))
        
    #First boundary conditon: the option payoff at t=maturity
    call_option_maturity = np.array([])
    for i in S:
        payoff_option = max(i-K,0)
        call_option_maturity = np.append(call_option_maturity, payoff_option)
    
    V[:, t_steps] = call_option_maturity[::-1]
    
    
    #Second boundary X=Xmax 
    discretized_time = np.linspace(0, T, t_steps+1)
    Smaxs = np.array([])
    for i in range(len(discretized_time)):
        Smax_new = Smax - K * np.exp(r*discretized_time[i])
        Smaxs = np.append(Smaxs,Smax_new)
    
    V[0,:] = Smaxs[::-1]
    

    if scheme == "FTCS":
        A = FTCS.spare_matA(dt, sigma, dx, r, x_steps)
        B = FTCS.spare_matB(x_steps)
    elif scheme == "CN":
        A = CN.spare_matA(dt, sigma, dx, r, x_steps)
        B = CN.spare_matB(dt, sigma, dx, r, x_steps)
    
    
    
    #####0 was -1 when it was correct#####
    for i in range(t_steps, 0, -1):
        #flip because other wise the V_n_plus_1 is in wrong order to solve with B       
        flipped_V = np.flip(V[:,i])       

        V_n_plus_1 = np.dot(B,flipped_V)
        #print(V_n_plus_1)
        new_V= np.linalg.solve(A,V_n_plus_1)
        
        #flip to put it right in total grid
        V[:,i-1]  = np.flip(new_V)
        #print(np.round(V,0))
    
    #print(X[int(x_steps/2)])
    index_closest_S0 = np.abs(S - S_0).argmin()
    #print(S[index_closest_S0 ])
    return V[index_closest_S0,0], V, S

    
S_0s = [100,110,120]
T = 1
sigma = 0.3
r = 0.04
K = 110
t_steps = 100 #Number of time steps
S_steps = 100 #Number of stock steps
dt = T/t_steps
scheme = 'FTCS'

for S_0 in S_0s:
    results = PDE(S_0, K, r, sigma, T, t_steps,S_steps,dt,scheme)
    price_V_S0 = results[0]
    prices_V = results[1]
    S_prices = results[2]

    #print(np.round(prices_V,1))

    print('Price option with S=100 and T=0 with numerical FCTS: ',price_V_S0)
    print('Price option with S=100 and T=0 with analytical solution: ',black_scholes.callBlackScholes(S_0,T, r, K, sigma)[0])

    np.arange(0, T+dt,dt)
    x_axis_labels = np.round(np.linspace(0, T, t_steps+1),2) # labels for x-axis
    y_axis_labels = np.round(S_prices[::-1], 0) # labels for y-axis

    df = pd.DataFrame(data=prices_V,columns=x_axis_labels, index=y_axis_labels)

    plot = sns.heatmap(df,cbar_kws={'label': 'Option price'})
    plt.title('Grid of solving the Black-Scholes PDE with the FTCS scheme')
    plot.set_xlabel("Time", fontsize = 20)
    plot.set_ylabel("Stock Price", fontsize = 20)

    """
    X, Y = np.meshgrid(x_axis_labels, y_axis_labels)

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, prices_V,rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')
    ax.set_xlabel('time')
    ax.set_ylabel('Stockprice')
    ax.set_zlabel('Option price')
    ax.view_init(30, 220)
    fig
    plt.show()
    """
    plt.show()

```

```python
#determine grid size with S0 is 100 with FTCS
scheme ='FTCS'
S_0 = 100
max_tsteps = 100
max_xsteps = 100
start_from = 2
error_matrix = np.zeros((max_tsteps-start_from,max_xsteps-start_from ))

for t_steps in range(start_from, max_tsteps):
    for x_steps in range(start_from , max_xsteps):
        dt = T/t_steps
        results = PDE(S_0, K, r, sigma, T, t_steps,x_steps,dt,scheme)
        analytical = black_scholes.callBlackScholes(S_0,T, r, K, sigma)[0]
        error = abs(results[0]-analytical)
        error_matrix[t_steps-start_from , x_steps-start_from ] = error
print(np.round(error_matrix,1))

df2 = pd.DataFrame(data=error_matrix,columns=np.arange(start_from ,max_xsteps), index=np.arange(start_from,max_tsteps))

plot = sns.heatmap(df2,cbar_kws={'label': 'Error'})
plt.title('Error depending on grid size')
plot.set_xlabel("Stock steps", fontsize = 10)
plot.set_ylabel("Time steps", fontsize = 10)

print(np.amin(error_matrix))
result1 = np.where(error_matrix == np.amin(error_matrix))
print(result1)
optimal_grid_size1 = result1
```

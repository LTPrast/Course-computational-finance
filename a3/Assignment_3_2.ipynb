```python
import numpy as np
import matplotlib.pyplot as plt
from utils import black_scholes, FTCS, CN
import seaborn as sns
import pandas as pd
import scipy.sparse.linalg as sp
from mpl_toolkits.mplot3d import Axes3D

sns.set_style("whitegrid")


def PDE(S_0, K, r, sigma, T, t_steps, x_steps,dt,scheme):
 
    S_max = 10000
    S_min = 1
    x_max = np.log(S_max)
    x_min = np.log(S_min)
    dx = (x_max - x_min)/x_steps
    
    #everly spaced logarithacly
    X = np.linspace(x_min, x_max, x_steps+1)
    #then add the the exponential
    S = np.exp(X)
    Smax = S[-1]
    

    #initialize grid and (at Xmin/Smin -> V=0)
    V = np.zeros((x_steps+1,t_steps+1))
        
    #First boundary conditon: the option payoff at t=maturity
    call_option_maturity = np.array([])
    for i in S:
        payoff_option = max(i-K,0)
        call_option_maturity = np.append(call_option_maturity, payoff_option)
    
    V[:, t_steps] = call_option_maturity[::-1]
    
    
    #Second boundary X=Xmax 
    discretized_time = np.linspace(0, T, t_steps+1)
    Smaxs = np.array([])
    for i in range(len(discretized_time)):
        Smax_new = Smax - K * np.exp(-r*discretized_time[i])
        Smaxs = np.append(Smaxs,Smax_new)
    
    V[0,:] = Smaxs[::-1]

    

    if scheme == "FTCS":
        A = FTCS.spare_matA(dt, sigma, dx, r, x_steps-1)[0]
        B = FTCS.spare_matB(x_steps-1)

    elif scheme == "CN":
        A = CN.spare_matA(dt, sigma, dx, r, x_steps-1)[0]
        B = CN.spare_matB(dt, sigma, dx, r, x_steps-1)
    
    up_diag = FTCS.spare_matA(dt, sigma, dx, r, x_steps-1)[1]
    down_diag = FTCS.spare_matA(dt, sigma, dx, r, x_steps-1)[2]
    
    correction_boundary = np.zeros(x_steps-1)
    correction_boundary[0] = down_diag * V[0,i]
    correction_boundary[-1] = up_diag * V[-1,i] 
        
    for i in range(t_steps, 0, -1):
        #flip because other wise the V_n_plus_1 is in wrong order to solve with B       
        flipped_V = np.flip(V[:,i])
        V_n_plus_1 = B.dot(flipped_V[1:-1])

        V_n= sp.spsolve(A, (V_n_plus_1 - correction_boundary[::-1]))
        
        #flip to put it right in total grid
        V[1:-1:,i-1]  = np.flip(V_n)
    
    index_closest_S0 = np.abs(S - S_0).argmin()
    

    option_price_S0 = np.interp(S_0, S, V[:,0][::-1])
    return option_price_S0, V, S,X,index_closest_S0, dx
    
T = 1
sigma = 0.3
r = 0.04
K = 110
```

```python
#determine grid size with S0 is 100 with FTCS
scheme ='FTCS'
S_0_1 = 100
max_tsteps = 1000
max_xsteps = 1000
start_from = 100
steps = 50
steps_t = int((max_tsteps-start_from)/steps)
steps_x = int((max_xsteps-start_from)/steps)
error_matrix = np.zeros((steps_t,steps_x))

for t_steps in range(start_from, max_tsteps, steps):
    print(t_steps)
    for x_steps in range(start_from , max_xsteps, steps):
        dtau = T/t_steps
        dt = -dtau
        results = PDE(S_0_1, K, r, sigma, T, t_steps,x_steps,dt,scheme)
        analytical = black_scholes.callBlackScholes(S_0_1,T, r, K, sigma)[0]
        error = abs(results[0]-analytical)
        error_matrix[int((t_steps-start_from)/steps) , int((x_steps-start_from)/steps)] = error

df1 = pd.DataFrame(data=error_matrix,columns=np.arange(start_from ,max_xsteps, steps), index=np.arange(start_from,max_tsteps, steps))

plot = sns.heatmap(df1,cbar_kws={'label': 'Error'}, vmin=0, vmax=0.01)
plt.title('Absolute error depending on grid size')
plot.set_xlabel("Stock steps", fontsize = 10)
plot.set_ylabel("Time steps", fontsize = 10)

print(np.amin(error_matrix))
result1 = np.where(error_matrix == np.amin(error_matrix))
print(result1)
optimal_grid_size1 = result1
```

```python
####forward difference for calculation of delta over every time and every stock price####
#needed for forward difference method in 3d

#shoulf for sure be -1 since with the forward method you cant calculate the delta of the highest I did 10 because at the 
#it started to act weird because of the boundary condition (this for every place where -1 is seen)
opt_prices8 = prices_V7[::-1]

stock_steps8 = opt_prices8.shape[0]
t_steps8 = opt_prices8.shape[1]
deltas8 = np.zeros((stock_steps8-10, t_steps8))

for i in range(t_steps8):
    for j in range(stock_steps8-10):
        delta = (opt_prices8[j+1][i]-opt_prices8[j][i])/(S_prices7[j+1] - S_prices7[j])
        #print(delta)
        deltas8[j,i] = delta

#to get a matrix of delta structed in the same way as the option price
#after transformation for  th forward FD method
deltas8 = deltas8[::-1]  

x_axis_labels = np.linspace(0, T, t_steps8)
y_axis_labels = X_prices7[:-10][::-1]

X, Y = np.meshgrid(x_axis_labels, y_axis_labels)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, deltas8,rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.set_title('Delta with forward difference method')
ax.set_xlabel('time')
ax.set_ylabel('X')
ax.set_zlabel('Delta')
ax.view_init(20, 200)
fig
plt.show()
```
